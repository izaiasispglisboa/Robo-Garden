```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier, MLPRegressor
from sklearn.metrics import accuracy_score, confusion_matrix, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder




# Carregar o dataset pré-processado
data_path = r"/Users/zaivi/Desktop/Visual Basic/Robo Garden/data_refined.csv"
df_class = pd.read_csv(data_path)

# Converter a coluna 'diagnosis' para valores numéricos
le = LabelEncoder()
df_class['diagnosis'] = le.fit_transform(df_class['diagnosis'])

# Separar atributos e alvo
X_class = df_class.drop('diagnosis', axis=1)
y_class = df_class['diagnosis']

# Verifique se todas as colunas de X_class são numéricas
print(X_class.dtypes)

# Dividir os dados em treino, validação e teste
X_train_val, X_test, y_train_val, y_test = train_test_split(
    X_class, y_class, test_size=0.1, random_state=42, stratify=y_class
)
X_train, X_val, y_train, y_val = train_test_split(
    X_train_val, y_train_val, test_size=0.1111, random_state=42, stratify=y_train_val
)

# (A) Classificação com Scikit-Learn MLPClassifier

mlp_clf = MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=42)
mlp_clf.fit(X_train, y_train)

y_pred = mlp_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print("Scikit-Learn MLP Classifier Results")
print(f"Accuracy: {accuracy * 100:.2f}%")
print("Confusion Matrix:")
print(cm)
print("\n(Note: Minimum required accuracy is 94%)\n")


# (B) Classificação com uma Rede Neural Keras
model_clf = Sequential([
    Dense(64, activation='relu', input_dim=X_train.shape[1]),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')  # Classificação binária
])

model_clf.compile(optimizer=Adam(learning_rate=0.001), 
                  loss='binary_crossentropy', 
                  metrics=['accuracy'])

# Treinamento do modelo
model_clf.fit(X_train, y_train, epochs=50, batch_size=32,
              validation_data=(X_val, y_val), verbose=0)

loss, accuracy_keras = model_clf.evaluate(X_test, y_test, verbose=0)
y_pred_keras = (model_clf.predict(X_test) > 0.5).astype("int32")
cm_keras = confusion_matrix(y_test, y_pred_keras)

print("Keras Neural Network Classifier Results")
print("-----------------------------------------")
print(f"Accuracy: {accuracy_keras * 100:.2f}%")
print("Confusion Matrix:")
print(cm_keras)

```
