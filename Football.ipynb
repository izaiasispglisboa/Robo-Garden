```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv(r"C:\Users\justino\Downloads\results.csv")

data = data.dropna() 

print(f"Number of tuples: {len(data)}")
print(f"Number of tournaments: {len(data['tournament'].unique())}")

data['date'] = pd.to_datetime(data['date'])

# Count matches played in 2018
matches_2018 = data[data['date'].dt.year == 2018]
print(f"Number of matches played in 2018: {len(matches_2018)}")

# Calculate home team results
home_team_results = data['home_score'] - data['away_score']
home_team_wins = (home_team_results > 0).sum()
home_team_losses = (home_team_results < 0).sum()
home_team_draws = (home_team_results == 0).sum()

# Visualize home team results
plt.figure(figsize=(8, 6))
plt.pie([home_team_wins, home_team_losses, home_team_draws], 
        labels=['Wins', 'Losses', 'Draws'], 
        autopct='%1.1f%%', 
        startangle=90)
plt.title('Home Team Results')
plt.show()

# Visualize neutral column
data['neutral'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Neutral Venue Matches')
plt.show()
```
